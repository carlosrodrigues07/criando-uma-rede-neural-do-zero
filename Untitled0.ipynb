# Análise Exploratória
Nas células podemos ver que o dataset é composto de campos numéricos
(número de adultos na reserva, número de noites na semana) e categóricos (tipo de
plano alimentar, tipo de quarto reservado).

```python
hotel_dataset = pd.read_csv('/content/Hotel Reservations.csv')
hotel_dataset.head()
```

Para podemos saber como cada reserva(cada linha do datset) se encaixa na faixa de preço, iremos criar uma nova coluna chamada label_avg_price_room, que vai servir como label (rótulo) para a classificaçã. A nova coluna contém 1 quando a coluna avg_price_per_room tiver valor menor ou igual a 85,  número 2 quando a coluna avg_price_per_room tiver valor maior que 85 e menor que 115 e o valor 3 se a coluna avg_price_per_room tiver valor maior ou igual a 115.

```python
# Criar a coluna de rótulos (label_avg_price_per_room)
hotel_dataset['label_avg_price_per_room'] = hotel_dataset['avg_price_per_room'].apply(
    lambda x: 1 if x <= 85 else (2 if 85 < x < 115 else 3)
)
hotel_dataset.head()
```

Ponto positivo é que não hpa campos incompletos ou inconsitentes, o que facilita o processo de contrução do modelo
